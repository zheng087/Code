```python
path = "openjian.txt"
f= open(path,"w")
f.write("hello\n")
f.close()
"""
r - 讀取(檔案要存在)

w - 新建檔案寫入(檔案可不存在)

a - 資料附加到就檔案後面(游標指在E0F)

r+ - 讀取舊資料並寫入(檔案需存在且游標指在開頭)

w+ - 清空檔案內容，新寫入的東西在讀出(檔案可不存在，會自行新增)

a+ - 資料附加到舊檔案後面(游標指在E0F)，可讀取資料
"""

f=open(path,"r")
print(f.read())
f.close()
```
