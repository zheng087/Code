```python
import random

rock = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
'''

paper = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)
'''

scissors = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
'''
image = [rock,paper,scissors] #圖片清單


mep=0
comp=0

while True:    
    mep=0
    comp=0
    while mep<2 and comp<2:
        me = int(input("0 , 1 , 2"))
        print(image[me])
        com = random.randint(0,2)
        print(image[com])
        if me==0 and com==2 or (me>com and me-com==1):
            print("win")
            mep+=1
        elif me==2 and com==1 or (me<com and com-me==1):
            print("lose")
            comp+=1
        else:
            print("draw")
    if mep>comp:
        print("你贏了")
    else:
        print("你輸了")
    re=input("隨便按一個鍵進入下一局,想結束按(N/n)")
    print(re)
    if re=="n" or re=="N":
        break
        
print("結束")
  

        
```
