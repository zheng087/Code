```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import openpyxl
wb=openpyxl.Workbook()
ws=wb.active

ws["A1"]="職稱"
ws["B1"]="公司"
ws["C1"]="地區"
ws["D1"]="薪水"
ws["E1"]="薪資下限"
ws["F1"]="薪資上限"
ws["G1"]="網址"

data = requests.get("https://www.1111.com.tw/search/job?ks=大數據&page=1")
soup = BeautifulSoup(data.text)
#soup.find_all("div",class_="job_item_info")!=[]:

page=1
while soup.find_all("div",class_="job_item_info")!=[]:
    print("正在讀取",page,"頁")
    page+=1
    for job in soup.find_all("div",class_="job_item_info"):
        a=job.h5.text
        b=job.h6.text
        c=job.find_all("a",class_="job_item_detail_location mr-3 position-relative")[0].text
        d=job.find(class_="job_item_detail_salary ml-3 font-weight-style digit_6").text
        d1=d[:2]
        e=job.a["href"]
        price = ""
        for word in d:
            if word == "." or word == "萬" or word == "~" or word.isdigit()==1:
                price+=word
        if "~" in price:
            low_price=price[:price.find("~")]
            high_price=price[price.find("~")+1:]
        else:
            low_price=price
            high_price=price
    
        if "萬" in low_price:
            low_price=float(low_price.replace("萬",""))*10000
        else:
            low_price=float(low_price)

        if "萬" in high_price:
            high_price=float(high_price.replace("萬",""))*10000
        else:
            high_price=float(high_price)  
        d = (low_price)
        dd =(high_price)
        
        #ws.append([a,b,c,d1,d,dd,e])  
        print(a,b,c,d,dd,e)
        #print("-----------------")
    data = requests.get("https://www.1111.com.tw/search/job?ks=大數據&page="+str(page))
    soup = BeautifulSoup(data.text)
sleep(2)

wb.save("test20221208.xlsx")

print("done")
```
