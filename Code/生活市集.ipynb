```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

option = webdriver.ChromeOptions() #獲取預設設定 
prefs = {"profile.default_content_setting_values":{"notifications":2}} #創造新設定
option.add_experimental_option("prefs",prefs) #更新設定

driver = webdriver.Chrome("/Users/jasonzheng/Downloads/Chromedriver.exe",options=option) #開始webdriver並更改預設設定
driver.get("https://www.buy123.com.tw/site/category/12/居家生活雜貨/#ref=d_main_category_12") #開啟網址
```

```python
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
page = 1
anchor = driver.find_elements(By.CLASS_NAME,("item-wrap"))
anchor1 = []
#item-img
for ii in range(1,1001):
    try:
        print("第"+str(page)+"頁")
        page+=1
        anchor1.append(anchor[ii].get_attribute("href"))
        #print(ccc["href"])
        if ii % 10 == 0:
            driver.execute_script("window.scrollTo(0,"+str((ii//10) * 3000)+")")
        sleep(1)
    except:
        pass
        print("error")

```
